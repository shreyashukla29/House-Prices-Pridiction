```python
data['PoolQC']=data['PoolQC'].fillna('none')
```

```python
data['MiscFeature']=data['MiscFeature'].fillna('none')
```

```python
data['Alley']=data['Alley'].fillna('none')
```

```python
data['Fence']=data['Fence'].fillna('none')
```

```python
data['FireplaceQu']=data['FireplaceQu'].fillna('none')
```

```python
data['LotFrontage']=data.groupby('Neighborhood')['LotFrontage'].apply(lambda x: x.fillna(x.median()))
```

```python
for col in ['GarageType','GarageFinish','GarageQual','GarageCond']:
    data[col]=data[col].fillna('none')
```

```python
for col in ['GarageYrBlt','GarageCars','GarageArea']:
    data[col]=data[col].fillna(0)
```

```python
for col in ['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2']:
    data[col]=data[col].fillna('none')
```

```python
data['MasVnrType']=data['MasVnrType'].fillna('none')
```

```python
data['MasVnrArea']=data['MasVnrArea'].fillna(0)
```

```python
m=data['MSZoning'].mode()[0]
m
```

```python
data['MSZoning']=data['MSZoning'].fillna(m)
```

```python
for col in ['BsmtHalfBath','BsmtFullBath','TotalBsmtSF','BsmtUnfSF','BsmtFinSF2','BsmtFinSF1','BsmtUnfSF','TotalBsmtSF']:
    data[col]=data[col].fillna(0)
```

```python
data['Utilities']=data['Utilities'].fillna('AllPub')
```

```python
data['Electrical']=data['Electrical'].fillna(data['Electrical'].mode()[0])
```

```python
data['Functional']=data['Functional'].fillna('Typ')
```

```python
data['SaleType']=data['SaleType'].fillna('WD')
data['Exterior2nd']=data['Exterior2nd'].fillna('VinylSd')
data['Exterior1st']=data['Exterior1st'].fillna('VinylSd')
```

```python
data['KitchenQual']=data['KitchenQual'].fillna(data['KitchenQual'].mode()[0])
data['GarageCars']=data['GarageCars'].fillna(0)
data['GarageArea']=data['GarageArea'].fillna(0)
```

```python
for cols in ['OverallQual','OverallCond','YearBuilt','YearRemodAdd','MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
       'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature',
       'SaleType', 'SaleCondition']:
    data[cols]=pd.Categorical(data[cols])
    data[cols]=data[cols].cat.codes
```
